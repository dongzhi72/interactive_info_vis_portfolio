```python
import pandas as pd; 

# 1. Load the dataset
file_name = "/Users/eloiseee/Documents/INFO474/HW4/interactive_info_vis_portfolio/data/earthquakes-datetime-id-2026-01-30_22-11-37_-0800_earthquakes-2026-01-30_22-11-37_-0800.tsv.csv"
df = pd.read_csv(file_name, encoding='utf-16', sep='\t')
df.head()
```

```python

# 2. Get the rows with the top 10 'deaths'
top_10_deaths = df.nlargest(10, 'Deaths')
```

```python
top_10_deaths
```

```python
# 3. Export to a new CSV file
# 'index=False' prevents pandas from adding an extra column for row numbers
output_file = "/Users/eloiseee/Documents/INFO474/HW4/interactive_info_vis_portfolio/data/top_10_deadliest_earthquakes.csv"
top_10_deaths.to_csv(output_file, index=False)

print(f"Success! Top 10 records exported to {output_file}")
```
